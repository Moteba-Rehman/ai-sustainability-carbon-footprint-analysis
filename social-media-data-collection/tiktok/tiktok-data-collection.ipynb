# TikTok Data Collection using Hashtags

This notebook collects public TikTok posts related to the following hashtags:

- #ghibliart
- #ghiblistyle
- #ghiblistudio
- #ghiblitrend

The collected data will be used for future analysis as part of the **GenAI Carbon Footprint Research Project**.

```python
import pandas as pd
import matplotlib.pyplot as plt
from apify_client import ApifyClient
import getpass
```

```python
APIFY_TOKEN = getpass.getpass("Enter Apify API Token: ")

client = ApifyClient(APIFY_TOKEN)

print("✅ Connected to Apify")
```

```python
hashtags = [
    "ghibliart",
    "ghiblistyle",
    "ghiblistudio",
    "ghiblitrend"
]

print("Hashtags to collect:")

for tag in hashtags:
    print("-", tag)
```

```python
run_input = {
    "hashtags": hashtags,
    "resultsPerPage": 100
}

print("Starting TikTok scraper...")

run = client.actor("clockworks/tiktok-hashtag-scraper").call(
    run_input=run_input
)

print("✅ Scraping completed.")
```

```python
dataset_id = run.default_dataset_id

items = list(
    client.dataset(dataset_id).iterate_items()
)

tiktok_df = pd.DataFrame(items)

print(f"Collected {len(tiktok_df)} posts.")

display(tiktok_df.head())
```

```python
tiktok_clean = tiktok_df[[
    "searchHashtag",
    "id",
    "text",
    "hashtags",
    "webVideoUrl",
    "diggCount",
    "commentCount",
    "shareCount",
    "playCount",
    "createTimeISO",
    "authorMeta",
    "locationMeta"
]].copy()

tiktok_clean.rename(columns={
    "searchHashtag": "Search Hashtag",
    "id": "Post ID",
    "text": "Caption",
    "hashtags": "Hashtags",
    "webVideoUrl": "Video URL",
    "diggCount": "Likes",
    "commentCount": "Comments",
    "shareCount": "Shares",
    "playCount": "Views",
    "createTimeISO": "Timestamp",
    "authorMeta": "Author",
    "locationMeta": "Location"
}, inplace=True)

tiktok_clean.to_csv(
    "tiktok_ghibli_posts.csv",
    index=False
)

print(f"Saved {len(tiktok_clean)} posts.")

display(tiktok_clean.head())
```

```python
print("Dataset Shape:")
print(tiktok_clean.shape)

print("\nColumns:")
print(tiktok_clean.columns.tolist())

print("\nMissing Values:")
print(tiktok_clean.isnull().sum())

tiktok_clean.describe(include="all")
```
