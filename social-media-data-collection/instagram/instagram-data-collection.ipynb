# Instagram Data Collection

## Objective

This notebook collects publicly available Instagram posts using a predefined list of hashtags. The collected data will be used for future analysis of social media trends related to Generative AI and digital content.

The notebook demonstrates a reproducible workflow for:

- Authenticating with Instagram
- Collecting posts using hashtags
- Extracting relevant metadata
- Exporting the collected dataset for further analysis

```python
from apify_client import ApifyClient
import pandas as pd
```

```python
import getpass

APIFY_TOKEN = getpass.getpass("Enter your Apify API Token: ")

client = ApifyClient(APIFY_TOKEN)

print("✅ Connected to Apify")
```

```python
hashtags = [
    "ghibliart",
    "ghiblistyle",
    "ghiblistudio",
    "ghiblitrend"
]

print("Hashtags to collect:")

for tag in hashtags:
    print("-", tag)
```

```python
run_input = {
    "directUrls": [
        "https://www.instagram.com/explore/tags/ghibliart/",
        "https://www.instagram.com/explore/tags/ghiblistyle/",
        "https://www.instagram.com/explore/tags/ghiblistudio/",
        "https://www.instagram.com/explore/tags/ghiblitrend/"
    ],
    "resultsType": "posts",
    "resultsLimit": 100
}

print("Starting Instagram scraper...")

run = client.actor("apify/instagram-api-scraper").call(
    run_input=run_input
)

print("Scraping completed!")
```

```python
dataset_id = run.default_dataset_id

items = list(
    client.dataset(dataset_id).iterate_items()
)

instagram_df = pd.DataFrame(items)

print(f"Collected {len(instagram_df)} posts.")

display(instagram_df.head())
```

```python
# Keep only useful columns

instagram_clean = instagram_df[[
    "inputUrl",
    "id",
    "shortCode",
    "caption",
    "hashtags",
    "url",
    "likesCount",
    "commentsCount",
    "timestamp",
    "ownerUsername",
    "ownerFullName",
    "locationName"
]].copy()

instagram_clean.rename(columns={
    "inputUrl": "Hashtag URL",
    "id": "Post ID",
    "shortCode": "Short Code",
    "caption": "Caption",
    "hashtags": "Hashtags",
    "url": "Post URL",
    "likesCount": "Likes",
    "commentsCount": "Comments",
    "timestamp": "Timestamp",
    "ownerUsername": "Username",
    "ownerFullName": "Full Name",
    "locationName": "Location"
}, inplace=True)

instagram_clean.to_csv(
    "instagram_ghibli_posts.csv",
    index=False
)

print(f"Saved {len(instagram_clean)} posts.")

display(instagram_clean.head())
```

```python
print("Dataset Shape:")
print(instagram_clean.shape)

print("\nColumns:")
print(instagram_clean.columns.tolist())

print("\nMissing Values:")
print(instagram_clean.isnull().sum())

instagram_clean.describe(include="all")
```
