```python
keywords = [
    "ghibliart",
    "ghiblistyle",
    "ghiblistudio",
    "ghiblitrend"
]
```

```python
all_videos = []

ydl_opts = {
    "quiet": True,
    "extract_flat": True,
}

for keyword in keywords:

    print(f"Collecting: {keyword}")

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:

        results = ydl.extract_info(
            f"ytsearch100:{keyword}",
            download=False
        )

        for video in results["entries"]:

            all_videos.append({
                "Keyword": keyword,
                "Video ID": video.get("id"),
                "Title": video.get("title"),
                "Uploader": video.get("channel"),
                "Upload Date": video.get("upload_date"),
                "Duration (sec)": video.get("duration"),
                "Views": video.get("view_count"),
                "Video URL": f"https://www.youtube.com/watch?v={video.get('id')}"
            })

print(f"\nCollected {len(all_videos)} videos.")
```

```python
youtube_df = pd.DataFrame(all_videos)

print(f"Collected {len(youtube_df)} videos.")

display(youtube_df.head())
```

```python
youtube_clean = youtube_df[[
    "Keyword",
    "Video ID",
    "Title",
    "Uploader",
    "Upload Date",
    "Duration (sec)",
    "Views",
    "Video URL"
]].copy()

youtube_clean.to_csv(
    "youtube_ghibli_posts.csv",
    index=False
)

print(f"Saved {len(youtube_clean)} videos.")

display(youtube_clean.head())
```

```python
print("Dataset Shape:")
print(youtube_clean.shape)

print("\nColumns:")
print(youtube_clean.columns.tolist())

print("\nMissing Values:")
print(youtube_clean.isnull().sum())

youtube_clean.describe(include="all")
```
